```python
import importlib
import matplotlib.pyplot as plt
import numpy as np
import os
import sys

path = sys.path[0]
path = os.path.join(path, "..")
if path not in sys.path:
    sys.path.insert(1, path)
```

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "fire.style"))

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)

viz = drawable.Drawable(plt.figure(figsize=(10, 5)))

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

"""
Instead of adding a legend, add a text at the end of the time series.
"""
viz.plot(range(10), line)
viz.text(9 + 0.15, line[-1], 'A', color='C0', fontweight=600, va='center')

viz.set_title('Timeline', loc='left', fontweight='bold', pad=25) # pad to make space for a subtitle
viz.text(0, 1, 'A simple time series.',
         alpha=0.8, ha='left', va='bottom', transform=plt.gca().transAxes)
plt.show()
```

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "fire.style"))

figure, axis = plt.subplots(2, 1, figsize=(10, 10))

"""
Draw the first axis.
"""
viz = drawable.Drawable(figure, axis[0])
viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)
viz.plot(range(10), line, color='C0')
viz.text(9 + 0.15, line[-1], 'A', color='C0', fontweight=600, va='center')

viz.set_title('Timeline', loc='left', fontweight='bold', pad=25) # pad to make space for a subtitle
viz.text(0, 1, 'A simple time series.',
         alpha=0.8, ha='left', va='bottom', transform=axis[0].transAxes)

"""
Draw the second axis.
"""

viz = drawable.Drawable(figure, axis[1])

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)
viz.plot(range(10), line, color='C1')
viz.text(9 + 0.15, line[-1], 'A', color='C1', fontweight=600, va='center')

viz.set_title('Timeline', loc='left', fontweight='bold', pad=25) # pad to make space for a subtitle
viz.text(0, 1, 'Another simple time series.',
         alpha=0.8, ha='left', va='bottom', transform=axis[1].transAxes)

# plt.subplots_adjust(hspace=0.5) # this may be needed
plt.show()
```
