# Styles

Multiplex comes with multiple styles to quickly customize the look and feel of your visualizations. This notebook is a quick reference to help you see how the different styles look with different visualizations.

Each style has one primary color `C0`, three highlight colors (`C1`, `C2` and `C3`) and a background color `C4`.

```python
%load_ext autoreload
%autoreload 2

import importlib
import matplotlib.pyplot as plt
import os
import sys

path = sys.path[0]
path = os.path.join(path, "..")
if path not in sys.path:
    sys.path.insert(1, path)

import examples
from multiplex import drawable
drawable = importlib.reload(drawable)
```

```python
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex.style"))
drawable = importlib.reload(drawable)
examples = importlib.reload(examples)

figure = plt.figure(figsize=(20, 30))
gs = figure.add_gridspec(3, 2)
examples.draw_timeseries(drawable.Drawable(figure, figure.add_subplot(gs[0, :])))
examples.draw_bar_100(drawable.Drawable(figure, figure.add_subplot(gs[2, :])),
                      { 'alpha': 0.2 })
examples.draw_text(drawable.Drawable(figure, figure.add_subplot(gs[1, 0])))
examples.draw_graph(drawable.Drawable(figure, figure.add_subplot(gs[1, 1])))
plt.savefig('exports/0-multiplex-style.png', dpi=300, bbox_inches='tight')
```

```python
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex-dark.style"))
drawable = importlib.reload(drawable)
examples = importlib.reload(examples)

figure = plt.figure(figsize=(20, 30))
gs = figure.add_gridspec(3, 2)
examples.draw_timeseries(drawable.Drawable(figure, figure.add_subplot(gs[0, :])))
examples.draw_bar_100(drawable.Drawable(figure, figure.add_subplot(gs[2, :])),
                      { 'alpha': 0.2 })
examples.draw_text(drawable.Drawable(figure, figure.add_subplot(gs[1, 0])))
examples.draw_graph(drawable.Drawable(figure, figure.add_subplot(gs[1, 1])))
plt.savefig('exports/0-multiplex-dark-style.png', dpi=300, bbox_inches='tight')
```

```python
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "modern.style"))
drawable = importlib.reload(drawable)
examples = importlib.reload(examples)

figure = plt.figure(figsize=(20, 30))
gs = figure.add_gridspec(3, 2)
examples.draw_timeseries(drawable.Drawable(figure, figure.add_subplot(gs[0, :])))
examples.draw_bar_100(drawable.Drawable(figure, figure.add_subplot(gs[2, :])),
                      { 'alpha': 0.2 })
examples.draw_text(drawable.Drawable(figure, figure.add_subplot(gs[1, 0])))
examples.draw_graph(drawable.Drawable(figure, figure.add_subplot(gs[1, 1])))
plt.savefig('exports/0-modern-style.png', dpi=300, bbox_inches='tight')
```

```python
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "modern-academic.style"))
drawable = importlib.reload(drawable)
examples = importlib.reload(examples)

figure = plt.figure(figsize=(20, 30))
gs = figure.add_gridspec(3, 2)
examples.draw_timeseries(drawable.Drawable(figure, figure.add_subplot(gs[0, :])))
examples.draw_bar_100(drawable.Drawable(figure, figure.add_subplot(gs[2, :])))
examples.draw_text(drawable.Drawable(figure, figure.add_subplot(gs[1, 0])))
examples.draw_graph(drawable.Drawable(figure, figure.add_subplot(gs[1, 1])),
                    edge_style={ 'color': 'C1' }, name_style={ 'color': 'C1' },
                    node_style={ 'color': 'C4', 'edgecolor': 'C1', 'linewidth': 2 })
plt.savefig('exports/0-modern-academic-style.png', dpi=300, bbox_inches='tight')
```
