# 1. Getting started

Multiplex is based on matplotlib. Anything that you can do with matplotlib, you can also do it with Multiplex. The biggest change you'll experience in Multiplex is that all functionality revolves around the Drawable class.

The Drawable class is very simple. It expects only a figure and the subplot where you will plot visualizations. However, anything that you normally do with a matplotlib visualization, you can also do with a Drawable.

### Why Multiplex?

Multiplex makes it easier to create beautiful visualizations, whether you want to explore or explain your data. The way it achieves this is by providing:

- Custom matplotlib styles;
- Functionality to caption visualizations, as well as annotate any visualization with text; and
- New types of visualizations not available in matplotlib, such as the graph and text-based visualizations.

You can view all of the styles in the [styles notebook](https://github.com/NicholasMamo/multiplex-plot/blob/master/examples/0.%20Styles.ipynb).

This notebook is a quickstart guide to using Multiplex, including how to add captions. The next notebooks will explain how to annotate visualizations and how to use the new visualization types introduced in Multiplex.

The first cell is boilerplate code that imports matplotlib and prepares to import Multiplex.

```python
%load_ext autoreload
%autoreload 2

import importlib
import matplotlib.pyplot as plt
import numpy as np
import os
import sys

path = sys.path[0]
path = os.path.join(path, "..")
if path not in sys.path:
    sys.path.insert(1, path)
```

To start using Multiplex, import the library. It is also recommended to use one of [Multiplex's styles](https://github.com/NicholasMamo/multiplex-plot/blob/master/examples/0.%20Styles.ipynb) so you can start creating beautiful visualizations immediately.

The most important thing there is to know about Multiplex is that everything revolves around the Drawable class. The Drawable class is a very simple, but very important thing.

The Drawable class wraps a matplotlib figure and the plot where Multiplex will draw all visualizations. When you instantiate a Drawable, you need to provide, at least, the figure. You can provide a subplot where to draw visualizations. If you do not provide one, the default subplot will be used.

In the rest of this notebook and to simplify examples, the same process will always be used:

1. Import all the necessary libraries and styles,
2. Load the data which the visualization will use,
3. Create a general style for the visualization,
4. Create specific styling for elements we want to highlight, and
5. Draw the visualization.

Not all steps are necessary, naturallyâ€”if we already loaded the libraries, there is no need to reload them. Similarly, we may choose to highlight no data.

In the first example, we follow this process in order and draw a very simple time series. **Note that all function calls go through the Drawable class.** This is because any function you can call on matplotlib subplots can also be called on the Drawable class.

```python
# Import the Drawable class and set the style
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex.style"))

# Generate the data
x = range(10)
y = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)

# Create a general style
style = { 'color': 'C1' }

# Draw the visualization
viz = drawable.Drawable(plt.figure(figsize=(10, 5)))
viz.plot(x, y, **style)

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')
viz.set_title('An unremarkable time series', loc='left')
plt.show()
```

That's all well and good, but Multiplex enhances matplotlib. In the next example, observe two of Multiplex's features:

1. We add a caption to the subplot to describe the data.
2. We use one of Multiplex's new visualization types: the time series. You can learn more about them in the `time series tutorial <https://github.com/NicholasMamo/multiplex-plot/blob/master/examples/3.%20Time%20Series.ipynb>`_.

```python
# Import the Drawable class and set the style
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex.style"))

# Generate the data
x = range(10)
y = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)

# Create a general style
style = { 'color': 'C1' }
label_style = { 'fontweight': 700 }

# Draw the visualization
viz = drawable.Drawable(plt.figure(figsize=(10, 5)))
viz.draw_time_series(x, y, label='A', label_style=label_style, **style)

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')
viz.set_title('An unremarkable time series', loc='left')
viz.set_caption('A simple time series.')
plt.show()
```

The style not to your liking? Multiplex has more styles to choose from. Choose one to your liking from the [styles notebook](https://github.com/NicholasMamo/multiplex-plot/blob/master/examples/0.%20Styles.ipynb).

```python
# Import the Drawable class and set the style
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex-dark.style"))

# Generate the data
x = range(10)
y = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)

# Create a general style
style = { 'color': 'C1' }
label_style = { 'fontweight': 700 }

# Draw the visualization
viz = drawable.Drawable(plt.figure(figsize=(10, 5)))
viz.draw_time_series(x, y, label='A', label_style=label_style, **style)

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')
viz.set_title('An unremarkable time series', loc='left')
viz.set_caption('A simple time series, but dark.')
plt.show()
```
