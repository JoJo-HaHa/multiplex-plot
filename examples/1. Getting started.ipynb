# 1. Getting started

Multiplex revolves around a Drawable class. The Drawable class is essentially made up of a figure and an axis, or a subplot. Anything that you can do with a subplot, you can do with a Drawable class.

```python
%load_ext autoreload
%autoreload 2

import importlib
import matplotlib.pyplot as plt
import numpy as np
import os
import sys

path = sys.path[0]
path = os.path.join(path, "..")
if path not in sys.path:
    sys.path.insert(1, path)
```

To use Drawable's functionality, create a Drawable instance, providing at least the figure. If you do not provide a subplot, the plot's default subplot will be used.

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "fire.style"))

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)

viz = drawable.Drawable(plt.figure(figsize=(10, 5)))

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

"""
Instead of adding a legend, add a text at the end of the time series.
"""
viz.plot(range(10), line)
viz.text(9 + 0.15, line[-1], 'A', color='C0', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
viz.set_caption('A simple time series.')
plt.show()
```

You can provide the subplot to the Drawable to draw on that particular axis. You can do this for every subplot that you have.

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "fire.style"))

figure, axis = plt.subplots(2, 1, figsize=(10, 10))

"""
Draw the first axis.
"""
viz = drawable.Drawable(figure, axis[0])
viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)
viz.plot(range(10), line, color='C0')
viz.text(9 + 0.15, line[-1], 'A', color='C0', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
viz.set_caption('A simple time series.')

"""
Draw the second axis.
"""

viz = drawable.Drawable(figure, axis[1])

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)
viz.plot(range(10), line, color='C1')
viz.text(9 + 0.15, line[-1], 'A', color='C1', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
caption = """  Another simple time series. The caption can be split into multiple lines. You can define the linebreaks yourself.
               However, if a line is too long, it wraps around automatically, like this one. Since it exceeds the axis length it starts over in a new line, and that line wraps around as well. Crazy!
"""
caption = """
    This is a multi-level
    caption.
"""
viz.set_caption(caption)

plt.subplots_adjust(hspace=0.7) # this may be needed to make space for the caption
plt.show()
```

The style not to your liking? Multiplex also has a dark style.

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "fire-dark.style"))

figure, axis = plt.subplots(1, 1, figsize=(10, 5))

"""
Draw the first axis.
"""
viz = drawable.Drawable(figure, axis)
viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

for (i, label) in zip(range(0, 3), ('A', 'B', 'C')):
    line = np.linspace(0.4, 0.3 + 0.15 * i, num=10) + np.random.normal(0, 0.03, 10)
    viz.plot(range(10), line, color=f'C{i}')
    viz.text(9 + 0.15, line[-1], label, color=f'C{i}', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
viz.set_caption('A simple time series using the dark style.')

plt.show()
```
