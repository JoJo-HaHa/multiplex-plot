# 1. Getting started

Multiplex does not re-invent matplotlib's wheel. The biggest change you'll experience when working with Multiplex is that all functionality revolves around a Drawable class.

The Drawable class is very simple. It expects only a figure and the subplot where you will plot visualizations. However, anything that you normally do with a matplotlib visualization, you can also do with a Drawable.

Why, then, use Multiplex? This notebook gives you a very simple tour of Multiplex, and how it can help you create better visualizations.

```python
%load_ext autoreload
%autoreload 2

import importlib
import matplotlib.pyplot as plt
import numpy as np
import os
import sys

path = sys.path[0]
path = os.path.join(path, "..")
if path not in sys.path:
    sys.path.insert(1, path)
```

To use Drawable's functionality, create a Drawable instance, providing at least the figure. If you do not provide a subplot, the plot's default subplot will be used.

The Drawable enhances matplotlib. For startes, it lets you add a caption to your visualizations to explain the visualization.

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex.style"))

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)

viz = drawable.Drawable(plt.figure(figsize=(10, 5)))

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

"""
Instead of adding a legend, add a text at the end of the time series.
"""
viz.plot(range(10), line, color='C1')
viz.text(9 + 0.15, line[-1], 'A', color='C1', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
viz.set_caption('A simple time series.')
plt.show()
```

You can provide the subplot to the Drawable to draw on that particular axis. If you have multiple plots on the same figure, you can draw separately on each by creating one Drawable for every subplot.

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex.style"))

figure, axis = plt.subplots(2, 1, figsize=(10, 10))

"""
Draw the first axis.
"""
viz = drawable.Drawable(figure, axis[0])
viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)
viz.plot(range(10), line, color='C1')
viz.text(9 + 0.15, line[-1], 'A', color='C1', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
viz.set_caption('A simple time series.')

"""
Draw the second axis.
"""

viz = drawable.Drawable(figure, axis[1])

viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

line = np.linspace(0.4, 0.5, num=10) + np.random.normal(0, 0.03, 10)
viz.plot(range(10), line, color='C1')
viz.text(9 + 0.15, line[-1], 'A', color='C1', fontweight=600, va='center')

viz.set_title('Time series', loc='left')
caption = """  Another simple time series on a separate subplot."""
viz.set_caption(caption, align='justify')

plt.subplots_adjust(hspace=0.7) # this may be needed to make space for the caption
plt.show()
```

The style not to your liking? Multiplex has more styles to choose from, including a dark style and a modern style.

```python
from multiplex import drawable
drawable = importlib.reload(drawable)
plt.style.use(os.path.join(sys.path[0], '..', 'styles', "multiplex-dark.style"))

figure, axis = plt.subplots(1, 1, figsize=(10, 5))

"""
Draw the first axis.
"""
viz = drawable.Drawable(figure, axis)
viz.set_xlim(-1, 11)
viz.set_ylim(0.25, 0.75)
viz.set_xlabel('Time')
viz.set_ylabel('Value')

for (i, label) in zip(range(0, 3), ('A', 'B', 'C')):
    line = np.linspace(0.4, 0.3 + 0.15 * i, num=10) + np.random.normal(0, 0.03, 10)
    viz.plot(range(10), line, color=('C1' if i == 0 else 'C0'))
    viz.text(9 + 0.15, line[-1], label, color=('C1' if i == 0 else 'C0'), fontweight=600, va='center')

viz.set_title('Time series', loc='left')
viz.set_caption('A simple time series using the dark theme.')

plt.savefig('exports/1-time-series.png', dpi=300, bbox_inches='tight')
```
